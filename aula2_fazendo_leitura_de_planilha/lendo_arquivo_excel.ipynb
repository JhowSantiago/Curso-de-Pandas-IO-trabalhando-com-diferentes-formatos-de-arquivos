# Importando bibliotecas

```python
import pandas as pd
```

# Obtendo o arquivo Excel através do link

```python
url = 'https://github.com/alura-cursos/Pandas/blob/main/emissoes_CO2.xlsx?raw=True'
```

# Fazendo a leitura de arquivos excel

```python
dados_co2 = pd.read_excel(url)
```

# Visualizando o Data Frame

```python
dados_co2.head()
```

# Verificando se existe mais paginas dentro da planilha

```python
pd.ExcelFile(url).sheet_names
```

# Parâmetros da funcao read_excel

```python
percapita = pd.read_excel(url, sheet_name='emissoes_percapita')
percapita.head()
```

```python
fontes = pd.read_excel(url, sheet_name='fontes')
fontes.head()
```

```python
# Definindo um intervalo de coluna do Excel que será lido
intervalo = pd.read_excel(url, sheet_name='emissoes_C02', usecols='A:D')
intervalo
```

```python
# Definindo um intervalo de coluna do Excel que será lido e quantidade de linhas
intervalo = pd.read_excel(url, sheet_name='emissoes_C02', usecols='A:D',nrows=10)
intervalo
```

# Salvando o arquivo em formato Excel

```python
percapita.to_excel('co2_percapita.xlsx', index=False)
```
